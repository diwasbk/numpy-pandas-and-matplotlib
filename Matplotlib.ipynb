Importing Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Line Graph

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
plt.plot(x, y, marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Graph')
plt.show()
```

Bar Plot

```python
labels = ["A", "B", "C"]
values = [10, 25, 40]
plt.bar(labels, values)
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

Pie Chart

```python
sizes = [30, 35, 20, 15]
labels = ["A", "B", "C", "D"]
plt.pie(sizes, labels = labels, autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
```

Donut Chart

```python
sizes = [30, 40, 30]
labels = ["A", "B", "C"]
plt.pie(sizes, labels=labels, autopct="%1.1f%%", wedgeprops = dict(width=0.3))
plt.title("Donut Chart")
plt.show()
```
